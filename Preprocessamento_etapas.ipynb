```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import img_to_array, load_img
from tensorflow.keras.applications.imagenet_utils import preprocess_input

img_path = '/content/drive/MyDrive/UFC/TCC/dataset_odonto_ufc/Dataset_Mask_Copia_Com_Itens_Excluidos_UFC/101_22A_M.jpg'
```

```python
# Etapa 1: Carregar imagem original
original = load_img(img_path)
original_array = img_to_array(original).astype(np.uint8)

# Etapa 2: Redimensionamento e normalização
resized = cv2.resize(original_array, (128, 128))
normalized = preprocess_input(resized.copy())  # Normalização como no seu pipeline
normalized_vis = cv2.normalize(resized, None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)  # Para visualização

# Etapa 3: Conversão para escala de cinza (se necessário)
gray = cv2.cvtColor(resized, cv2.COLOR_RGB2GRAY)

# Plotagem
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(original_array.astype(np.uint8))
plt.title('Original')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(normalized_vis)
plt.title('Redimensionada e Normalizada')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(gray, cmap='gray')
plt.title('Convertida para Escala de Cinza')
plt.axis('off')

plt.tight_layout()
plt.show()
```
